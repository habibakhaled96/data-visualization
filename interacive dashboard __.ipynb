```python
df['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))'].fillna(df.groupby('Entity')['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))'].transform('mean'),inplace= True)
```

```python
df.dropna(subset=['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))','GDP per capita, PPP (constant 2017 international $)'])
```

```python
df= df.dropna(subset=['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))', 'GDP per capita, PPP (constant 2017 international $)'], how='all')
```

```python
df['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))'].fillna(method = 'bfill', inplace= True)
```

```python
df['GDP per capita, PPP (constant 2017 international $)'].fillna(method = 'ffill', inplace= True)
```

```python
df['Total population (Gapminder, HYDE & UN)']=df['Total population (Gapminder, HYDE & UN)'].fillna(method = 'bfill')
```

```python
df['Continent']=df['Continent'].fillna(method = 'bfill')
```

```python
df = df[(df["Year"] > 2000) ]
```

```python
df.drop('Code', inplace = True, axis=1)
```

```python
df = df[(df['Entity']!='World') & (df['Year']< 2015)]
```

```python
df=df.dropna()
```

```python
df = df.sort_values('Year', ascending = True)
```

```python
#title="Children Percentage(13-15) who reported being bullied %.1f " % avg_builled)

app = dash.Dash(external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])

fig = px.scatter(df,x= 'GDP per capita, PPP (constant 2017 international $)', y= 'Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))', size='Total population (Gapminder, HYDE & UN)', color= 'Continent',hover_name='Entity',title= 'Violence aganist Children Vs. GDP per capita', 
           log_x = True, template = 'presentation', animation_frame='Year', animation_group='Entity', size_max=50 ,trendline="ols",trendline_scope="overall" )
fig.update_layout(
    xaxis_title="GDP per capita",
    yaxis_title="Percentage")


avg_builled = df1['Percentage'].mean()
fig2 = px.choropleth(df1, locations="Code", color="Percentage",
                    color_continuous_scale=px.colors.cmocean.balance,
                    color_continuous_midpoint=avg_builled,
                    title="Children Percentage(13-15) who reported being bullied")


app.layout = html.Div(style = {
  'backgroundColor': '#111155'
}, children = [
   
    html.H1(
    children = 'Child Abuse',
    style = {
      'textAlign': 'center',
      'color': '#FFFFFF',
        'font-size': 50
    }
  ),
    
    
    html.Div(children=[
        
        html.Div([dbc.Card(dbc.CardBody([html.H5("physical punishment in Egypt", className="card-title",style={'textAlign':'center','fontSize':'150%'}),html.P("9%",style={'textAlign':'center','fontSize':'250%'}),]), color="#FFFFFF", inverse=True)],className="four columns", style={'display': 'inline-block','height': 200,'width':400,'left':20}),
       

        html.Div([dbc.Card(dbc.CardBody([html.H5("psychological punishment in Egypt", className="card-title",style={'textAlign':'center','fontSize':'150%'}),html.P("9%",style={'textAlign':'center','fontSize':'250%'}),]), color="#FFFFFF", inverse=True)], className="four columns", style={'display': 'inline-block','position': 'absolute', 'height': 200,'width':400 ,'top': 90,'left':440}),
        
        
        html.Div([dbc.Card(dbc.CardBody([html.H5("Both punishment in Egypt", className="card-title",style={'textAlign':'center','fontSize':'150%'}),html.P("73%",style={'textAlign':'center','fontSize':'250%'}),]), color="#FFFFFF", inverse=True)], className="four columns", style={'display': 'inline-block','position': 'absolute', 'height': 200,'width':400,'top': 90,'left':950}),
        
        
        
    ], className="row",style={'width': '100%', 'display': 'inline-block'}),

    
    html.Div(children=[
        
        html.Div([ 
                                html.Div(["Input Year: ", dcc.Input(id='input-year', value='2013', 
                                type='number', style={'height':'15px', 'font-size': 15}),], 
                                style={'font-size': 20, 'color': '#FFFFFF'}),
                                
                                html.Div(dcc.Graph(id='line-plot')),
                                ],className="six columns", style={'display': 'inline-block'}),
       

        html.Div([
            
            dcc.Graph(id='g2', figure=fig2)
        ], className="six columns", style={'display': 'inline-block','position': 'absolute', 'height': 100, 'top': 324}),
        
        
    ], className="row",style={'width': '100%', 'display': 'inline-block'}),
    
    
    html.Div( html.Div([
            
            dcc.Graph(id='g1', figure=fig)
        ], )
, className='row')])

# add callback decorator
@app.callback( Output(component_id='line-plot', component_property='figure'),
               Input(component_id='input-year', component_property='value'))

# Add computation to callback function and return graph
def get_graph(entered_year):
    # Select 2019 data
   
    if ((entered_year > 2004) and (entered_year < 2016)):
        df =  df_vi[df_vi['Year']==int(entered_year)]
    else:   
         df =  df_vi[df_vi['Year']==2014]
    
    # Group the data by Month and compute average over arrival delay time.
    #line_data = df.groupby('Month')['ArrDelay'].mean().reset_index()

    fig = go.Figure(data=go.Bar(y=df['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))'], x=df['Entity'], text= df['Percentage of children (2–14) who experience any violent discipline (UNICEF Global Databases (2016))']))
    fig.update_layout(title='Percentage of children (2–14) who experience any violent discipline', yaxis_title='Percentage', xaxis_title='Country')
    return fig

if __name__ == '__main__':
    app.run_server(debug = False)
```
